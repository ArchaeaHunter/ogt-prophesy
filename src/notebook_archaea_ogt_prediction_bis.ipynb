```python
import pandas as pd
import numpy as np
from toolbox import PredictorOGT
from ml_pipelines import models 
from sklearn.model_selection import KFold
```

# Data clean-up

```python
data_col = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']
target_col = ["OGT"]

df = pd.read_csv("../data/amino-acids-counts-archaea.tsv",sep="\t",index_col=0) 
# Convert amino-acids counts to frequences 
for aa in data_col : 
    df[aa] = df[aa]/df["aa_sum"]
# Create group column by genus for CV
df["genus"] = df["SpHierarchie"].str.split("~").str[4]
df
```

# Vizualisation of data

```python
data = df[data_col]
target = df[target_col]
model = models["baseline"]
cv = KFold(n_splits=5,shuffle=True,random_state=42)
working_folder = "results"
exp_name = "amino-acids-counts-archaea"
```

```python
pogt = PredictorOGT(data,target,model,cv,None,exp_name)
pogt.run_visualisation_data()
```

# Prediction with baseline

```python
pred_target = pogt.predict_target()
```

```python
pred_target
```

```python
pogt.print_error_estimation()
```

# RidgeCV model

```python
model = models["ridge"]
pogt = PredictorOGT(data,target,model,cv,None,exp_name)
pred_target = pogt.predict_target()
pogt.print_error_estimation()
pogt.permutation_importance_test()
```

# Spline Nystroem Ridge model

```python
model = models["ridge_spline_nystroem"]
pogt = PredictorOGT(data,target,model,cv,None,exp_name)
param_grid = {"spline__n_knots":[int(value) for value in np.linspace(5,50,10)],
    "spline__degree": [int(value) for value in np.linspace(10,100,10)],
    "nystroem__gamma": np.linspace(0,0.2,9),
    "nystroem__n_components": [int(value) for value in np.linspace(10,100,10)]}

pogt.search_hyperparams(param_grid,False)
pred_target = pogt.predict_target()
pogt.print_error_estimation()
pogt.permutation_importance_test()
```

# Random Forest

```python
model = models["random_forest"]
pogt = PredictorOGT(data,target,model,cv,None,exp_name)
param_grid = {"max_features": [1, 2, 3, 5, None],
            "max_leaf_nodes": [10, 100, 1000, None],
            "min_samples_leaf": [1, 2, 5, 10, 20, 50, 100]}
pogt.search_hyperparams(param_grid,False)
pred_target = pogt.predict_target()
pogt.print_error_estimation()
pogt.permutation_importance_test()
```

# Histogram Gradient Boosting

```python
from scipy.stats import loguniform

model = models["hist_gradient_boosting_regressor"]
pogt = PredictorOGT(data,target,model,cv,None,exp_name)
param_grid = {"max_iter": [3, 10, 30, 100, 300, 1000],
                "max_leaf_nodes": [2, 5, 10, 20, 50, 100],""
                "learning_rate": loguniform(0.01, 1)}
pogt.search_hyperparams(param_grid,False)
pred_target = pogt.predict_target()
pogt.print_error_estimation()
pogt.permutation_importance_test()
```
